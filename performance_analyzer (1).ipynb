```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")

# Plot: Gender vs Average Score
df['average'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
sns.boxplot(x='gender', y='average', data=df)
plt.title("Average Score by Gender")
plt.show()

# Plot: Parental Level of Education vs Average Score
plt.figure(figsize=(10,6))
sns.boxplot(x='parental level of education', y='average', data=df)
plt.xticks(rotation=45)
plt.title("Effect of Parental Education on Performance")
plt.show()

# Plot: Study time proxy (test preparation course)
sns.boxplot(x='test preparation course', y='average', data=df)
plt.title("Impact of Test Preparation Course on Scores")
plt.show()
```

## 📊 Summary of Insights

- Female students scored higher on average in reading and writing.
- Students who completed a test preparation course performed better across all subjects.
- Higher parental education is positively associated with better student performance.
- Math scores have a wider range, indicating greater variation among students.
